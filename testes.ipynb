```python
import requests

endpoint = "https://api.lyrics.ovh/v1/the weeknd/die for you"

response = requests.get(endpoint)
response.status_code
```

```python
print(response.json()['lyrics']) # exibe apenas o conteúdo na chave "lyrics". o print() formata as quebras de linhas

# json() = mostra conteúdo json da requisição
# json().keys() = mostra as chaves dentro do dicionário (o Python retorna arquivos JSON como dicionário)
```
